```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Check normality for each column in the dataframe
normality_results = {}

for column in df_reduced.columns:
    stat, p_value = stats.shapiro(df_reduced[column])
    normality_results[column] = {'statistic': stat, 'p_value': p_value}

# Display the normality test results
normality_results_df = pd.DataFrame(normality_results).T
normality_results_df
```

```python
def f_test(sample1, sample2):
    f = np.var(sample1, ddof=1)/np.var(sample2, ddof=1)  # Calculate F statistic
    df1 = len(sample1) - 1  # Degrees of freedom for sample1
    df2 = len(sample2) - 1  # Degrees of freedom for sample2
    p_value = stats.f.sf(f, df1, df2)  # Two-tailed p-value
    return f, p_value

# For the first pair, SEXVAR and PHYSHLTH, we need to clean the data first:
# - Exclude rows where PHYSHLTH is coded as 88 (which may represent not applicable or missing data)
# - Then, split PHYSHLTH into two groups based on SEXVAR

# Clean PHYSHLTH data
physhlth_clean = df_reduced.loc[df_reduced['PHYSHLTH'] != 88, 'PHYSHLTH']

# Split the data based on SEXVAR (assuming 1 for male and 2 for female)
male_physhlth = physhlth_clean[df_reduced['SEXVAR'] == 1]
female_physhlth = physhlth_clean[df_reduced['SEXVAR'] == 2]

# Perform the F-test for SEXVAR and PHYSHLTH
f_statistic_sexvar_physhlth, p_value_sexvar_physhlth = f_test(male_physhlth, female_physhlth)

# For the second pair, HLTHPLN1 and POORHLTH, we need to clean POORHLTH as well:
# - Exclude rows with NaN values in POORHLTH
# - Split POORHLTH into two groups based on HLTHPLN1

# Clean POORHLTH data
poorhlth_clean = df_reduced.loc[df_reduced['POORHLTH'].notna(), 'POORHLTH']

# Split the data based on HLTHPLN1 (assuming 1 for with and 2 for without healthcare coverage)
with_hc_poorhlth = poorhlth_clean[df_reduced['HLTHPLN1'] == 1]
without_hc_poorhlth = poorhlth_clean[df_reduced['HLTHPLN1'] == 2]

# Perform the F-test for HLTHPLN1 and POORHLTH
f_statistic_hlthpln1_poorhlth, p_value_hlthpln1_poorhlth = f_test(with_hc_poorhlth, without_hc_poorhlth)

# For the third pair, MARITAL and MENTHLTH, we need to decide how to group the MARITAL categories
# For simplicity, we'll compare those who are married (1) vs. not married (any other category)

# Clean MENTHLTH data
menthlth_clean = df_reduced.loc[df_reduced['MENTHLTH'].notna(), 'MENTHLTH']

# Split the data based on MARITAL
married_menthlth = menthlth_clean[df_reduced['MARITAL'] == 1]
not_married_menthlth = menthlth_clean[df_reduced['MARITAL'] != 1]

# Perform the F-test for MARITAL and MENTHLTH
f_statistic_marital_menthlth, p_value_marital_menthlth = f_test(married_menthlth, not_married_menthlth)

print(f"Sex vs Physical Health F-stats is {f_statistic_sexvar_physhlth} and p-value is {p_value_sexvar_physhlth}")
print(f"Health vs Poor Health F-stats is {f_statistic_hlthpln1_poorhlth} and p-value is {p_value_hlthpln1_poorhlth}")
print(f"Mental Health vs Marital F-stats is {f_statistic_marital_menthlth} and p-value is {p_value_marital_menthlth}")
```
