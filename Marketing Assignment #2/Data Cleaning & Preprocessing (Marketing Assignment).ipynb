```python
import pandas as pd
df = pd.read_csv(r"E:\Marketing Assignment\People Survey\responses.csv")
df.head(10)
```

```python
df.columns
```

```python
#Extracting only necessary columns
new_order = ['Age', 'Education', 'Gender', 'Smoking', 'Alcohol','Energy levels', 'Life struggles', 
             'Number of friends', 'Happiness in life']

new_df = df[new_order].head(250)
new_df.columns
```

```python
new_df.head(250)
```

```python
specific_columns = ['Age', 'Education', 'Gender', 'Smoking', 'Alcohol','Energy levels', 'Life struggles', 
                    'Number of friends', 'Happiness in life']

# Loop through each specified column and print unique values and their counts
for column in specific_columns:
    if column in df.columns:
        unique_values = df[column].unique()
        value_counts = df[column].value_counts()
        print(f"\nColumn: {column}")
        print(f"Unique values: {unique_values}")
        print(f"Frequency of unique values:\n{value_counts}")
    else:
        print(f"\nColumn: {column} is not present in the DataFrame")
```

```python
#Dropping all missing values and cleaning data
new_df = new_df.dropna()
```
